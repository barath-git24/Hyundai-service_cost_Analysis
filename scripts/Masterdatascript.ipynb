```python
import pandas as pd

df_customers = pd.read_csv("Datasets/customers.csv")
df_sales = pd.read_csv("Datasets/sales.csv")
df_service = pd.read_csv("Datasets/service.csv")
df_vehicle = pd.read_csv("Datasets/vehicle.csv")
```

```python
df_sales=df_sales.sort_values("sale_date")
```

```python
df_sales=df_sales.drop_duplicates(subset="vehicle_id",keep="last")
```

```python
df_sales.shape
```

```python
df_master=df_service.copy()
```

```python
df_master=df_master.merge(df_vehicle,on="vehicle_id",how="left")
```

```python
df_master=df_master.merge(df_sales,on="vehicle_id",how="left")
```

```python
df_master=df_master.merge(df_customers,on="customer_id",how="left")
```

```python
df_master.shape
```

```python
df_master.isnull().sum()
```

```python
df_master=df_master.dropna(subset=["customer_id"])
```

```python
df_master['dealer_region'] = df_master['dealer_region_x'].fillna(df_master['dealer_region_y'])

df_master.drop(columns=['dealer_region_x', 'dealer_region_y'], inplace=True)
```

```python
df_master['sale_date'] = pd.to_datetime(df_master['sale_date'])
df_master['service_date'] = pd.to_datetime(df_master['service_date'])
```

```python
# extract years
df_master['sale_year'] = df_master['sale_date'].dt.year
df_master['service_year'] = df_master['service_date'].dt.year

# age at sale
df_master['vehicle_age_at_sale'] = df_master['sale_year'] - df_master['purchase_year']

# age at service (MOST IMPORTANT)
df_master['vehicle_age'] = df_master['service_year'] - df_master['purchase_year']
```

```python
df_master['profit_estimate'] = df_master['final_price'] - df_master['discount']
```

```python
df_master['has_service'] = df_master['service_cost'].notna().astype(int)
```
