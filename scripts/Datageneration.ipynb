```python
import pandas as pd
import numpy as np
from faker import Faker
import random
```

```python
fake=Faker()
```

Customer_Table

```python
num_customers=5000
customer_data=[]
```

```python
for i in range(1,num_customers +1):
    customer_data.append({"customer_id":i,"name":fake.name(),"age":random.randint(20,70),"gender":random.choice(["Male","Female"]),"city":fake.city(),"income":random.randint(200000,2000000),"credit_score":random.randint(300,850)})
df_customers=pd.DataFrame(customer_data)
```

```python
df_customers.to_csv("customers.csv",index=False)
```

Vehicle_Table

```python
vehicle_models = [
    "i10", "i20", "Santro", "Grand i10 Nios",
    "Verna", "Aura", "Elantra",
    "Creta", "Tucson", "Venue", "Kona", "Alcazar",
    "Kona Electric", "Ioniq", "Ioniq 5"
]

# Segment mapping (REALISTIC)
segment_mapping = {
    "i10": "Hatchback",
    "i20": "Hatchback",
    "Santro": "Hatchback",
    "Grand i10 Nios": "Hatchback",

    "Verna": "Sedan",
    "Aura": "Sedan",
    "Elantra": "Sedan",

    "Creta": "SUV",
    "Tucson": "SUV",
    "Venue": "SUV",
    "Kona": "SUV",
    "Alcazar": "SUV",

    "Kona Electric": "Electric",
    "Ioniq": "Electric",
    "Ioniq 5": "Electric"
}

# Engine options per model
engine_mapping = {
    "Kona Electric": ["Electric"],
    "Ioniq": ["Electric"],
    "Ioniq 5": ["Electric"],
}

default_engines = ["Petrol", "Diesel"]

num_vehicles = 2000
vehicle_data = []

for i in range(1, num_vehicles + 1):
    model = random.choice(vehicle_models)
    segment = segment_mapping[model]

    # Engine logic
    if model in engine_mapping:
        engine_type = random.choice(engine_mapping[model])
    else:
        engine_type = random.choice(default_engines)

    # Realistic base price by segment
    if segment == "Hatchback":
        base_price = random.randint(500000, 900000)
    elif segment == "Sedan":
        base_price = random.randint(900000, 1500000)
    elif segment == "SUV":
        base_price = random.randint(1000000, 2500000)
    else:  # Electric
        base_price = random.randint(2000000, 5000000)

    vehicle_data.append({
        "vehicle_id": i,
        "model_name": model,
        "engine_type": engine_type,
        "base_price": base_price,
        "segment": segment,
        "purchase_year": random.randint(2015, 2022)
    })

df_vehicles = pd.DataFrame(vehicle_data)
```

```python
df_vehicles.to_csv("vehicle.csv",index=False)
```

Sales_Table

```python
num_sales = 10000
dealer_regions = ["Chennai", "Bangalore", "Hyderabad", "Mumbai", "Delhi"]
payment_types = ["Cash", "Loan", "EMI"]

sales_data = []

for i in range(1, num_sales + 1):
    customer_id = random.randint(1, num_customers)
    vehicle_id = random.randint(1, num_vehicles)
    base_price = df_vehicles.loc[df_vehicles['vehicle_id'] == vehicle_id, 'base_price'].values[0]
    discount = random.randint(0, 200000)
    final_price = base_price - discount
    sales_data.append({
        "sale_id": i,
        "customer_id": customer_id,
        "vehicle_id": vehicle_id,
        "sale_date": fake.date_between(start_date='-5y', end_date='today'),
        "final_price": final_price,
        "discount": discount,
        "dealer_region": random.choice(dealer_regions),
        "payment_type": random.choice(payment_types)
    })

df_sales = pd.DataFrame(sales_data)
```

```python
df_sales.to_csv("sales.csv",index=False)
```

service_Table

```python
service_types = ["Minor", "Major", "Repair"]
num_services = 50000  
service_data = []

for i in range(1, num_services + 1):
    vehicle_id = random.randint(1, num_vehicles)
    purchase_year = df_vehicles.loc[df_vehicles['vehicle_id'] == vehicle_id, 'purchase_year'].values[0]
    vehicle_age = 2026 - purchase_year
    service_type = random.choice(service_types)
    
    # Base cost depending on type and vehicle age
    cost = 5000
    if service_type == "Major":
        cost += 7000
    elif service_type == "Repair":
        cost += 3000
    cost += vehicle_age * 500  # older cars cost more
    if df_vehicles.loc[df_vehicles['vehicle_id'] == vehicle_id, 'segment'].values[0] == "SUV":
        cost += 2000
    cost += random.randint(-500, 500)
    
    service_data.append({
        "service_id": i,
        "vehicle_id": vehicle_id,
        "service_date": fake.date_between(start_date='-3y', end_date='today'),
        "service_type": service_type,
        "service_cost": cost,
        "warranty_claim": random.choice(["Yes", "No"]),
        "dealer_region": random.choice(dealer_regions)
    })

df_service = pd.DataFrame(service_data)
```

```python
df_service.to_csv("service.csv",index=False)
```
