```python
df_service = df_service.drop_duplicates(subset=['SERVICE_ID'])
df_vehicles = df_vehicles.drop_duplicates(subset=['VEHICLE_ID'])
```

```python
df_service['SERVICE_DATE'] = pd.to_datetime(df_service['SERVICE_DATE'])
df_vehicles['PURCHASE_YEAR'] = df_vehicles['PURCHASE_YEAR'].astype(int)
df_service['SERVICE_COST'] = df_service['SERVICE_COST'].astype(float)
```

```python
df_service.dtypes
```

```python
df_service['service_type'] = df_service['service_type'].str.strip().str.title()
df_service['warranty_claim'] = df_service['warranty_claim'].str.strip().str.title()
df_vehicles['engine_type'] = df_vehicles['engine_type'].str.strip().str.title()
```

```python
df_ml = df_service.merge(df_vehicles,on='VEHICLE_ID',how='left')
```

```python
df_ml.shape
```

```python
current_year = 2026
df_ml['VEHICLE_AGE'] = current_year - df_ml['PURCHASE_YEAR']
```

```python
df_ml.head(10)
```

```python
df_ml['WARRANTY_FLAG'] = df_ml['WARRANTY_CLAIM'].map({True: 1,False: 0})
```

```python
df_ml.head()
```

```python
df_ml.shape
```

```python
y=df_ml["SERVICE_COST"]
```

```python
X = df_ml[[
    
        'VEHICLE_AGE','WARRANTY_FLAG','BASE_PRICE','SERVICE_TYPE','ENGINE_TYPE','SEGMENT','DEALER_REGION']]
    
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
numeric_features = ['VEHICLE_ID','WARRANTY_FLAG', 'BASE_PRICE']

categorical_features = [ 'SERVICE_TYPE', 'ENGINE_TYPE', 'SEGMENT', 'DEALER_REGION']
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ],
    remainder='passthrough'
)
```

```python
X_processed = preprocessor.fit_transform(X)
```

```python
print(X_processed.shape)
```

```python
X_processed
```

```python
type(X_processed)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_processed,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
model=RandomForestRegressor(n_estimators=100,random_state=42,max_depth=10)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
print(y_pred[:5])
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mae=mean_absolute_error(y_test,y_pred)
rmse=mean_squared_error(y_test,y_pred)
r2=r2_score(y_test,y_pred)
```

```python
print("mae:",mae)
print("rmse:",rmse)
print("r2:",r2)
```
